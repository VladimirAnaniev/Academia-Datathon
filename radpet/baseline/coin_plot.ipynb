```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import numpy as np

EPS = 1e-10
def mape(y_true, y_pred):
    return np.mean(np.abs((y_true + EPS - y_pred) / y_true + EPS)) * 100
```

```python
coins = pd.read_csv('../SELECTED_COINS.csv').values.reshape((20,))
```

```python
av_mape = []
for coin in coins:
#     y_true = pd.read_csv('../data/train/'+str(coin)+'.csv', index_col='time')
    y_true = pd.read_csv('../data/test/'+str(coin)+'.csv', index_col='time')
#     y_true = y_true[10001:]
    y_true = y_true[1:]
    
#     y_pred = pd.read_csv('./'+str(coin)+'/checkpoints/y_pred_val.csv')
#     y_pred = pd.read_csv('./only_price/'+str(coin)+'/y_pred_test.csv')
    y_pred = pd.read_csv('./only_price_on_full/'+str(coin)+'/checkpoints/y_pred_val.csv')
    y_pred.index = y_true.index
    y_pred['true_price'] = y_true['price']
    mape_score = mape(y_true=y_pred['true_price'], y_pred=y_pred['price'])
    av_mape.append(mape_score)
    print('########{} and mape {} #######'.format(coin, mape_score))
    print(y_pred.head())
    print((y_pred['price'].values == y_true['price'].values).sum())
    print(y_pred['price'].describe())
    print()
    print(y_pred['true_price'].describe())
    plt.figure(figsize=(10,10));
    y_pred.plot()
    plt.show()
    print()
```

```python
np.array(av_mape).mean()
```
